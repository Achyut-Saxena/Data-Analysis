```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('covid-19-states-daily.csv', parse_dates=['date'])
df.head()
```

```python
pd.unique(df['state'])
```

```python
df.info()
```

```python
# bar graph
br = df[df['total']>=500]
fig = plt.figure(figsize = (30,20))
plt.bar(br['state'], br['total'])
plt.show()
```

```python
#box plot
bx = df[df['positive']>=1000]
df.plot.box()
```

```python
# line graph
ln = df[df['state']=='WA']
fig = plt.figure(figsize = (10,10))
plt.plot(ln['date'], ln['positive'], 'ro', label = 'Positive')
plt.plot(ln['date'], ln['negative'], 'g^', label = 'Negative')
plt.plot(ln['date'], ln['total'], label = 'Total')
plt.legend()
plt.show()
```

```python
# scatter plot
ln = df[df['state']=='CA']
fig = plt.figure(figsize = (10,10))
plt.scatter(ln['date'], ln['death'], label = 'Death')
plt.scatter(ln['date'], ln['total'], label = 'Total')
plt.scatter(ln['date'], ln['positive'], label = 'Positive')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
# single line graph
ln = df[df['state']=='CA']
fig = plt.figure(figsize = (10,10))
plt.plot(ln['date'], ln['death'], label = 'Death')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
# pie chart
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend()
plt.show() 
```
